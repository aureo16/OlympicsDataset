```python
import pandas as pd
import json
```

## 120 anos

```python
atletas = pd.read_csv("atletas.csv")
edicoes = pd.read_csv("edicoes.csv")
paises = pd.read_csv("comites.csv")
esportesModalidades = pd.read_csv("esportes.csv")
participacaoComites = pd.read_csv("participacaoComites.csv")
participacaoAtletas = pd.read_csv("participacaoAtletas.csv")
comiteDosAtletas = pd.read_csv("comiteDosAtletas.csv")
esportesDasEdicoes = pd.read_csv("esportesDasEdicoes.csv")
```

```python
edicoesInfo=[]
for k in range(len(edicoes)):
  ano = edicoes.loc[k].Ano

  edicao = edicoes.loc[edicoes.Ano==ano]
  participacaoAtletasEdicao = participacaoAtletas.loc[participacaoAtletas.AnoEdicao==ano]
  participacaoComitesEdicao = participacaoComites.loc[participacaoComites.AnoEdicao==ano]
  esportesEdicao = esportesDasEdicoes.loc[esportesDasEdicoes.AnoEdicao==ano]

  atletasInfo = []

  atletasEdicao = participacaoAtletasEdicao[['IdAtleta']]
  atletasEdicao = atletasEdicao.drop_duplicates("IdAtleta")
  atletasEdicao = atletasEdicao.reset_index(drop=True)

  for i in range(len(atletasEdicao)):
    id = atletasEdicao.loc[i].IdAtleta
    atleta = atletas.loc[id]
    if (atleta.AnoDeNascimento == "-"):
      ano = ""
    else:
      ano = str(int(float(atleta.AnoDeNascimento)))

    if (atleta.isna().Sexo):
      sexo = ""
    else:
      sexo = atleta.Sexo

    comiteAtleta = comiteDosAtletas.loc[comiteDosAtletas.IdAtleta==atleta.Id].IdComite.values[0]

    esportesAtleta = participacaoAtletasEdicao.loc[participacaoAtletasEdicao.IdAtleta==atleta.Id]
    esportesAtleta = esportesAtleta.reset_index(drop=True)
    esportesInfo = []
    for j in range(len(esportesAtleta)):
      participacao = esportesAtleta.loc[j]
      esporte = esportesModalidades.loc[esportesModalidades.Id==participacao.IdModalidade]
      esportesInfo.append({
          "nomeEsporte": esporte.EsportePai.values[0],
          "nomeModalidade": esporte.Nome.values[0],
          "peso": participacao.Peso,
          "altura": participacao.Altura,
          "medalha": participacao.Medalha
      })
    
    atletasInfo.append({
        "id": str(atleta.Id),
        "nome": atleta.Nome,
        "anoDeNascimento": ano,
        "sexo": sexo,
        "comite": comiteAtleta,
        "esportes": esportesInfo
    })

  participacaoComitesEdicao = participacaoComitesEdicao.reset_index(drop=True)
  comitesInfo = []
  for i in range(len(participacaoComitesEdicao)):
    comite = participacaoComitesEdicao.loc[i]

    pais = paises.loc[paises.Id==comite.IdComite]["País"].values[0]

    comitesInfo.append({
        "sigla": comite.IdComite,
        "país": pais,
        "qtdAtletas": str(comite.QtdAtletas),
        "qtdOuro": str(comite.QtdOuro),
        "qtdPrata": str(comite.QtdPrata),
        "qtdBronze": str(comite.QtdBronze),
        "classificacao": comite.Classificacao
    })

  esportesInfo = []
  esportesPai = esportesModalidades.loc[esportesModalidades.EsportePai=="-"]
  esportesPai = esportesPai.reset_index(drop=True)

  for i in range(len(esportesPai)):
    esporte = esportesPai.loc[i]
    modalidades = esportesModalidades.loc[esportesModalidades.EsportePai==esporte.Nome]
    modalidades = modalidades.reset_index(drop=True)

    modalidadesInfo = []
    for j in range(len(modalidades)):
      modalidade = modalidades.loc[j]
      info = esportesEdicao.loc[esportesEdicao.IdModalidade==modalidade.Id]
      if (len(info)!= 0):
        modalidadesInfo.append({
            "nome": modalidade.Nome,
            "ouro": info.Ouro.values[0],
            "prata": info.Prata.values[0],
            "bronze": info.Bronze.values[0]
        })
    
    if (len(modalidadesInfo)!=0):
      esportesInfo.append({
          "nome": esporte.Nome,
          "modalidades": modalidadesInfo
      })
  
  edicoesInfo.append({
    "ano": str(edicao.Ano.values[0]),
    "numeroDaEdicao": str(edicao.NumerodaEdicao.values[0]),
    "cidadeSede": edicao.Cidade.values[0],
    "totalDeAtletas": str(edicao.TotaldeAtletas.values[0]),
    "mascote": edicao.Mascote.values[0],
    "atletasParticipantes": atletasInfo,
    "comitesParticipantes": comitesInfo,
    "esportesParticipantes": esportesInfo
  })
```

## Conversão JSON

```python
modeloDocumento = {
    "edicoes": edicoesInfo
}
```

```python
with open("olimpiadas.json","w") as json_file:
  json.dump(modeloDocumento, json_file)
```

```python
ano = 2020

edicao = edicoes.loc[edicoes.Ano==ano]
participacaoAtletasEdicao = participacaoAtletas.loc[participacaoAtletas.AnoEdicao==ano]
participacaoComitesEdicao = participacaoComites.loc[participacaoComites.AnoEdicao==ano]
esportesEdicao = esportesDasEdicoes.loc[esportesDasEdicoes.AnoEdicao==ano]

atletasInfo = []
for i in range(len(atletas)):
  atleta = atletas.loc[i]
  if (atleta.AnoDeNascimento == "-"):
    ano = ""
  else:
    ano = str(int(float(atleta.AnoDeNascimento)))

  if (atleta.isna().Sexo):
    sexo = ""
  else:
    sexo = atleta.Sexo

  comiteAtleta = comiteDosAtletas.loc[comiteDosAtletas.IdAtleta==atleta.Id].IdComite.values[0]

  esportesAtleta = participacaoAtletasEdicao.loc[participacaoAtletasEdicao.IdAtleta==atleta.Id]
  esportesAtleta = esportesAtleta.reset_index(drop=True)
  esportesInfo = []
  for j in range(len(esportesAtleta)):
    participacao = esportesAtleta.loc[j]
    esporte = esportesModalidades.loc[esportesModalidades.Id==participacao.IdModalidade]
    esportesInfo.append({
        "nomeEsporte": esporte.EsportePai.values[0],
        "nomeModalidade": esporte.Nome.values[0],
        "peso": participacao.Peso,
        "altura": participacao.Altura,
        "medalha": participacao.Medalha
    })
  
  atletasInfo.append({
      "id": str(atleta.Id),
      "nome": atleta.Nome,
      "anoDeNascimento": ano,
      "sexo": sexo,
      "comite": comiteAtleta,
      "esportes": esportesInfo
  })

participacaoComitesEdicao = participacaoComitesEdicao.reset_index(drop=True)
comitesInfo = []
for i in range(len(participacaoComitesEdicao)):
  comite = participacaoComitesEdicao.loc[i]

  pais = paises.loc[paises.Id==comite.IdComite]["País"].values[0]

  comitesInfo.append({
      "sigla": comite.IdComite,
      "país": pais,
      "qtdAtletas": str(comite.QtdAtletas),
      "qtdOuro": str(comite.QtdOuro),
      "qtdPrata": str(comite.QtdPrata),
      "qtdBronze": str(comite.QtdBronze),
      "classificacao": comite.Classificacao
  })

esportesInfo = []
esportesPai = esportesModalidades.loc[esportesModalidades.EsportePai=="-"]
esportesPai = esportesPai.reset_index(drop=True)

for i in range(len(esportesPai)):
  esporte = esportesPai.loc[i]
  modalidades = esportesModalidades.loc[esportesModalidades.EsportePai==esporte.Nome]
  modalidades = modalidades.reset_index(drop=True)

  modalidadesInfo = []
  for j in range(len(modalidades)):
    modalidade = modalidades.loc[j]
    info = esportesEdicao.loc[esportesEdicao.IdModalidade==modalidade.Id]
    modalidadesInfo.append({
        "nome": modalidade.Nome,
        "ouro": info.Ouro.values[0],
        "prata": info.Prata.values[0],
        "bronze": info.Bronze.values[0]
    })
  
  esportesInfo.append({
      "nome": esporte.Nome,
      "modalidades": modalidadesInfo
  })
  
edicoesInfo.append({
  "ano": str(edicao.Ano.values[0]),
  "numeroDaEdicao": str(edicao.NumerodaEdicao.values[0]),
  "cidadeSede": edicao.Cidade.values[0],
  "totalDeAtletas": str(edicao.TotaldeAtletas.values[0]),
  "mascote": edicao.Mascote.values[0],
  "atletasParticipantes": atletasInfo,
  "comitesParticipantes": comitesInfo,
  "esportesParticipantes": esportesInfo
})
```

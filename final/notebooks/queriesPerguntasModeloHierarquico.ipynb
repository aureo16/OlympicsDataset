```python
import json
with open("olimpiadas.json", "r") as json_file:
    dados = json.load(json_file)
print(type(dados))
```

##Consultas

```python
#Pergunta: Qual o número de atletas por país em uma determinada olimpíada?

#Selecionando a olimpíada (Através do ano):
ano="1936"

for j in range (len(dados["edicoes"])):
  if (dados["edicoes"][j]["ano"]==ano):
    numero=j
    break

for i in range (len(dados["edicoes"][numero]["comitesParticipantes"])):
  print(dados["edicoes"][numero]["comitesParticipantes"][i]["país"]+": "+dados["edicoes"][numero]["comitesParticipantes"][i]["qtdAtletas"]+" atletas")

```

```python
#Pergunta: Quais foram os países ganhadores de medalha de ouro no esporte X nas últimas 5 olimpíadas?

#Selecionando o esporte (Através do nome):
esporte="Judo"
modalidade="Judo Women's Heavyweight"
edicoes=["2016", "2012", "2008", "2004", "2000"]

for i in range (5):
  ano=edicoes[i]
  for j in range (len(dados["edicoes"])-1, -1, -1):
    if (dados["edicoes"][j]["ano"]==ano):
      for k in range (len(dados["edicoes"][j]["esportesParticipantes"])):
        if (dados["edicoes"][j]["esportesParticipantes"][k]["nome"]==esporte):
          for l in range (len(dados["edicoes"][j]["esportesParticipantes"][k]["modalidades"])):
            if (dados["edicoes"][j]["esportesParticipantes"][k]["modalidades"][l]["nome"]==modalidade):
              print(dados["edicoes"][j]["ano"]+": "+dados["edicoes"][j]["esportesParticipantes"][k]["modalidades"][l]["ouro"])
              break   
```

```python
#Pergunta: Quais os atletas participantes de uma determinada olimpíada e suas respectivas informações?

#Selecionando a olimpíada (Através do ano):
ano="1936"

for j in range (len(dados["edicoes"])):
  if (dados["edicoes"][j]["ano"]==ano):
    numero=j
    break

for i in range (len(dados["edicoes"][numero]["atletasParticipantes"])):
  print(dados["edicoes"][numero]["atletasParticipantes"][i])

```

```python
#Pergunta: Quais as modalidades realizadas pelo atleta X na olimpíada X?

#Selecionando a olimpíada (Através do ano):
ano="2004"
#Selecionando o atleta (Através do nome):
atleta="Daiane Garcia dos Santos"

for j in range (len(dados["edicoes"])):
  if (dados["edicoes"][j]["ano"]==ano):
    numero=j
    break

for k in range (len(dados["edicoes"][numero]["atletasParticipantes"])):
  if (dados["edicoes"][numero]["atletasParticipantes"][k]["nome"]==atleta):
    for l in range (len(dados["edicoes"][numero]["atletasParticipantes"][k]["esportes"])):
      print(dados["edicoes"][numero]["atletasParticipantes"][k]["esportes"][l]["nomeModalidade"])
```

```python
#Pergunta: Quais os comitês participantes de uma determinada olimpíada e seus respectivos desempenhos?

#Selecionando a olimpíada (Através do ano):
ano="2012"

for j in range (len(dados["edicoes"])):
  if (dados["edicoes"][j]["ano"]==ano):
    numero=j
    break

for i in range (len(dados["edicoes"][numero]["comitesParticipantes"])):
  print(dados["edicoes"][numero]["comitesParticipantes"][i])

```

```python
#Pergunta: Quais os esportes de uma determinada olimpíada e os resultados de pódio?

#Selecionando a olimpíada (Através do ano):
ano="2012"

for j in range (len(dados["edicoes"])):
  if (dados["edicoes"][j]["ano"]==ano):
    numero=j
    break

for k in range (len(dados["edicoes"][numero]["esportesParticipantes"])):
    for l in range (len(dados["edicoes"][numero]["esportesParticipantes"][k]["modalidades"])):
      print(dados["edicoes"][numero]["esportesParticipantes"][k]["modalidades"][l])
```
